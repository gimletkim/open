```python
import re

f = open('/content/drive/MyDrive/BME_class/2022-2_3학기/의생명정보학개론/20221110-20221117_실습자료/lab_result_sample.csv', 'r', encoding='CP949')
rdr = csv.reader(f)
ff = open('/content/drive/MyDrive/BME_class/2022-2_3학기/의생명정보학개론/20221110-20221117_실습자료/L3908.csv', 'w', newline='')
wr=csv.writer(ff)
for x in rdr:                                            
        lab_code=x[2]
        lab_result=x[4]
        if lab_code=="L3098":                        
                m=re.match(r'(\d+)',lab_result) 
                if m:
                        wr.writerow(x)                
                else:
                        mm=re.search(r'(\d+)\D+(\d+)',lab_result)  
                        if mm:
                                x[4]=mm.group(2)                
                                wr.writerow(x)
                        else:
                                pass
ff.close()
f.close()
```
