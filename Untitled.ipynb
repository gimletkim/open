# minmax norm


```python
from sklearn.preprocessing import minmax_scale

minmax_norm_cor = [] 
for i in range(len(templates)):
    #print(i)
    tmpl_MinMax_scaled = minmax_scale(templates.loc[i], axis=0, copy=True)
    minmax_norm_cor.append(tmpl_MinMax_scaled)
    #print(tmpl_MinMax_scaled)
```

```python
# dist base ì§€ì  ë§Œë“¤ê¸°

dist_base = [] 
for i in range(len(minmax_norm_cor.loc[0])):
    dist_tmp = np.mean(minmax_norm_cor[i])
    dist_base.append(dist_tmp)
```

```python
# Way2: to get similarity using Euclidean distance
from sklearn.metrics.pairwise import euclidean_distances

data_points = []
euclidean_dist_column = [] 
for i, ts in enumerate(tmp_ts):
    for data in range(len(minmax_norm_cor[0])):
        euclidean_dist = euclidean_distances([[minmax_norm_cor[i][data], ts], [dist_base[i], ts]])
        euclidean_dist = euclidean_dist[1][0]
        euclidean_dist_column.append(euclidean_dist)
    globals()['dt_{}'.format(i)] = euclidean_dist_column
    data_points.append(globals()['dt_{}'.format(i)])
    euclidean_dist_column = []
```
